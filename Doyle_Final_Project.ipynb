## Rotten Tomatoes Business and Data Understanding 

### This data is all text data retrieved from Rotten Tomatoes. The authors of these reviews all work for a big publication company of some sort and are familiar with the show. We are assuming that they have seen all of the seasons leading up to season 8, as most people have. What we are trying to see here is how this season compares to that of season 5. Season 5 is a very important season because after season 5, the writers of the TV show will not have any source material to go off of. In this analysis we are only collecting the review data. There is not enough ratings on the show to make it worth while to scrape, we also do not need to know who is writing about the show.
```python
openfile = open("data/df8.csv", "r")
r = csv.reader(openfile)
reviews = []
for i in r:
    reviews.append(i)
openfile.close()

#remove the extra bracket
reviews = [x for y in reviews for x in y]
# Remove useless numbers and alphanumerical words
documents = [re.sub("[^a-zA-Z]+", " ", document) for document in reviews]
# tokenize
texts = [[word for word in document.lower().split() ] for document in documents]

#Bing-Liu lexicon (positive word list & negative word list)
positive_words = open("C:\\Users\\Grant\\Desktop\\670\\3_Sentiment_Analysis\\data\\positive-words.txt", "r").read().splitlines()
negative_words = open("C:\\Users\\Grant\\Desktop\\670\\3_Sentiment_Analysis\\data\\negative-words.txt", "r").read().splitlines()

for sentence in texts:
    positive = 0
    negative = 0
    for word in sentence:
        if word in positive_words:
            positive += 1
        elif word in negative_words:
            negative += 1
    #print(positive, negative, positive-negative, "Positive" if positive-negative > 0 else "Neutral" if positive-negative == 0 else "Negative")
    
positive_review = []
negative_review = []
neutral_review = []

for sentence in texts:
    positive = 0
    negative = 0
    for word in sentence:
        if word in positive_words:
            positive += 1
        elif word in negative_words:
            negative += 1
    if positive-negative > 0:
        positive_review.append(sentence)
    elif positive-negative == 0:
        neutral_review.append(sentence)
    else:
        negative_review.append(sentence)
        

print(len(positive_review))
print(len(negative_review))
print(len(neutral_review))
```
